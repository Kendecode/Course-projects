```python
# importing neccessary dependencies 
from flask import Flask, render_template, request
from pymongo import MongoClient
# ObjectId function is used to convert the id string to an objectid that MongoDB can understand
from bson.objectid import ObjectId

# web application instantiation
app = Flask(__name__)

# Create the mongodb client
client = MongoClient('localhost', 27017)

db = client.DataForm # creating your flask database using your mongo client 
DataForm = db.DataForm # creating a collection called "DataForm"

@app.route("/")
def user_form():
    return render_template("user_form.html")

@app.route("/submit", methods=["POST"])
def submit_data():
# Extract user data and expenses from the form         
    user = {
        "age" : request.form.get("age"),
        "gender" : request.form.get("gender"),
        "income" : request.form.get("income"),
        "expenses" : {
            "utilities": request.form.get("utilities"),
            "entertainment": request.form.get("entertainment"),
            "school_fees": request.form.get("school_fees"),
            "shopping": request.form.get("shopping"),
            "healthcare": request.form.get("healthcare")
        }
    }
    DataForm.insert_one(user)

    return "Data submitted successfully! <br> <a href='/'>Go back to form</a>"


if __name__ == "__main__":
    
    app.run()

```

```python
import csv

class User:
    def __init__(self, age, gender, income, expenses):
        self.age = age
        self.gender = gender
        self.income = income
        self.utilities = expenses["utilities"]
        self.entertainment = expenses["entertainment"]
        self.school_fees = expenses["school_fees"]
        self.shopping = expenses["shopping"]
        self.healthcare = expenses["healthcare"]
        
def get_users():
    return DataForm.find()

def export_data(data):
    # List to store User objects
    users = []

    # Loop through collected data and create User objects
    for user_data in data:
        user = User(user_data["age"], user_data["gender"], user_data["income"], user_data["expenses"])
        users.append(user)

    # Export data to CSV file
    with open("user_data.csv", "w") as csvfile:
        fieldnames = ["age", "gender", "income", "utilities", "entertainment", "school_fees", "shopping", "healthcare"]
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        for user in users:
            user_dict = user.__dict__
            writer.writerow(user_dict)

# retrieve data from MongoDB DataForm
users_list = get_users()

export_data(users_list)
```
